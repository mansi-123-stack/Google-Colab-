```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Data
df = pd.read_csv("Regulatory Affairs of Road Accident Data 2020 India.csv")

# Step 3: Data Cleaning
df['Count'].fillna(0, inplace=True)

# Step 4: EDA
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# A. Total Accidents by City
city_accidents = df.groupby('Million Plus Cities')['Count'].sum().sort_values(ascending=False)
plt.figure()
sns.barplot(x=city_accidents.values, y=city_accidents.index, palette='viridis')
plt.title('Total Road Accidents by City')
plt.xlabel('Total Accidents')
plt.ylabel('City')
plt.tight_layout()
plt.show()

# B. Accidents by Cause Category
cause_category = df.groupby('Cause category')['Count'].sum().sort_values(ascending=False)
plt.figure()
sns.barplot(x=cause_category.values, y=cause_category.index, palette='rocket')
plt.title('Accidents by Cause Category')
plt.xlabel('Total Accidents')
plt.ylabel('Cause Category')
plt.tight_layout()
plt.show()

# C. Accidents by Cause Subcategory
subcategory = df.groupby('Cause Subcategory')['Count'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 10))
sns.barplot(x=subcategory.values, y=subcategory.index, palette='mako')
plt.title('Accidents by Cause Subcategory')
plt.xlabel('Total Accidents')
plt.ylabel('Subcategory')
plt.tight_layout()
plt.show()

# D. Outcomes of Incidents
outcomes = df.groupby('Outcome of Incident')['Count'].sum().sort_values(ascending=False)
plt.figure()
sns.barplot(x=outcomes.index, y=outcomes.values, palette='Set2')
plt.title('Road Accident Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# E. Heatmap: Cause Category vs Outcome
pivot = df.pivot_table(index='Cause category', columns='Outcome of Incident', values='Count', aggfunc='sum', fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Cause Category vs Outcome')
plt.tight_layout()
plt.show()
```

### Conclusion:
- **Top Accident-Prone Cities**: Mumbai, Delhi, Bengaluru
- **Major Causes**: Road Features and Traffic Violations
- **Frequent Subcategories**: Straight Road, Curved Road, Use of Mobile Phone
- **Common Outcomes**: Minor injuries and grievous injuries
- **Recommendation**: Focus on improving road design and enforcing traffic rules
